```python
# Fazendo a requisição para a página de notícias
url = 'https://www.ons.org.br/paginas/imprensa/noticias'
response = requests.get(url)
html = response.text

# Criando o objeto Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')


# Buscando todos os elementos que contêm os títulos das notícias
bloco_noticias = soup.find_all('div', class_='noticia')
site_tag = soup.find('meta', property= 'og:title')
site_nome = site_tag['content']


print(f'Órgão: {site_nome} | Atualizado em : {datetime.now()} ')
print('Última atualização local: ', datetime.now())
print(f"Quantidade de notícias encontradas: {len(bloco_noticias)}")
print("-" * 50)

for i in range(len(bloco_noticias)):

    noticias = bloco_noticias[i]
    
    info_tag = noticias.find('div', class_='info')
    info = info_tag.find('a')
    titulo = info.text.strip()
    link = info['href'].strip()

    data_div = noticias.find('div', class_='data')
    dia = data_div.find('p').text.strip() 
    mes = data_div.find('span').text.strip()
    data_publicacao = f'{dia}/{mes}'
    
        
    print(f"Data: {data_publicacao}")
    print(f"Título: {titulo}")
    print(f'Link: {link}')
    print("-" * 50)

```

```python
import time
from datetime import datetime
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Configura navegador headless
options = Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
driver = webdriver.Chrome(options=options)

driver.get('https://www.ons.org.br/paginas/imprensa/noticias')
time.sleep(3)  # espera que o JS carregue os blocos de notícias

soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

bloco_noticias = soup.select('div.list-noticias > div.noticia')
print(f"Noticias encontradas: {len(bloco_noticias)}")

for noticia in bloco_noticias:
    # extrai dia/mês
    data_div = noticia.find('div', class_='data')
    dia = data_div.find('p').text.strip() if data_div else '-'
    mes = data_div.find('span').text.strip() if data_div else '-'
    data = f"{dia}/{mes}"

    # extrai título e link
    a = noticia.find('div', class_='info').find('a')
    titulo = a.text.strip() if a else '—'
    link = a['href'] if a and a.has_attr('href') else '—'

    print(f"{data}  |  {titulo}")
    print(f"→ {link}")
    print('-'*40)
```

```python
# Fazendo a requisição para a página de notícias
url = 'https://www.epe.gov.br/pt/imprensa/noticias'
response = requests.get(url)
html = response.text
url_base = 'https://www.epe.gov.br'

# Criando o objeto Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# Buscando todos os elementos que contêm os títulos das notícias
site_tag = soup.find('img', {'title': True})
site_nome = site_tag['title']
bloco_noticias = soup.find_all('div', class_ = 'item')

print(f'Órgão: {site_nome} | Atualizado em : {datetime.now()} ')
print('Última atualização local: ', datetime.now())
print("-" * 50)

for i in range(5):
    noticias = bloco_noticias[i]
    
    titulo_tag = noticias.find('h2', class_='blue')
    titulo = titulo_tag.text.strip()
    tag_area = noticias.find('a', class_='tag-area')
    subtitulo = tag_area.text.strip()
    data_tag = noticias.find('span', class_='date')
    data_publicacao = data_tag.text.strip()
    link_tag = noticias.find('a')
    href = link_tag['href'].strip()
    link = url_base + href if href.startswith('/') else href if href else 'Link não encontrado'
    
        
    print(f"Data: {data_publicacao}")
    print(f"Categoria: {subtitulo}")
    print(f"Título: {titulo}")
    print(f'Link: {link}')
    print("-" * 50)

```
